# Customer Segmentation Notebook

## Quick Note
Started this notebook to revisit clustering concepts. Added markdown comments for clarity.


```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from utils import normalize

# Loading dataset
df = pd.read_csv('customer_data.csv')
df.head()
```

## Thought Process
Doing a basic EDA before clustering.


```python
# Quick EDA
df.describe()
```

## Clustering Section
Using KMeans with 3 clusters.


```python
# Normalizing selected columns
cols = ['annual_income', 'spending_score', 'visits_per_month']
df_norm = normalize(df, cols)

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_norm[cols])

df.head()
```
